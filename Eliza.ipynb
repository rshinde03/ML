```python
#function is used to replace first person with second person in the input text
def replacewords(response, reflections):
    for key in reflections.keys():
        if response == key:
            return reflections.get(key)
    return response
```

```python
#function to convet the text to lower case and split the text. It calls the functions
#to remove punctutions and replacing the words
def transformationwithreplacedword(response):
    response= remove_punctuations(response)
    response= response.lower()
    words=re.split(r'[\s\t\n]+', response)
    new_str=[]
    for word in words:
        new_str.append(replacewords(word, reflections))
        
    transformedstr= ' '.join(new_str)
    return transformedstr
```

```python
#function that searches in the dictionary for the the regular expression and returns the response 
def respond(response, expdictionary):
    answer="Can you tell more about it?"#default answer
    
    if re.match(r'Yes|yes', response):
        answer = "You seem quite sure. Elaborate please."
        
    elif re.match(r'No|no', response):
        answer = "Why do you deny? Elaborate please."
            
    elif re.match(r'Maybe|maybe', response):
        answer = "What's the reason for the uncertainity?"               
            
         
    else:
        for x in expdictionary:  
            xl = re.compile(x, re.IGNORECASE)
            is_match = re.match(xl, response)
            if is_match:
                answer = random.choice(expdictionary[x])
                position = answer.find('%')
                while position > -1:
                    count = int(answer[position+1:position+2])
                    answer = answer[:position] + \
                    transformationwithreplacedword(is_match.group(count)) + \
                    answer[position+2:]
                    position = answer.find('%')
                 
    return answer
```
